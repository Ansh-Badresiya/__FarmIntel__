### Split dataset for Crop Cetegory Prediction dataset

```python
crop_df = df[['State','District','Season','Year','Crop','Crop_Category']]
crop_df.sample(5)
```

```python
crop_df[crop_df['Crop_Category'].isna()]
```

```python
import os
from sklearn.model_selection import train_test_split

# ── Output directory ────────────────────────────────────────────────
PROCESSED_DIR = '../data/processed'
os.makedirs(PROCESSED_DIR, exist_ok=True)

# ── crop_df  (target: Crop) ──────────────────────────────────────────
crop_df_clean = crop_df.dropna(subset=['Crop_Category']).reset_index(drop=True)

crop_train, crop_test = train_test_split(
    crop_df_clean, test_size=0.2, random_state=42
)

crop_train.to_csv(os.path.join(PROCESSED_DIR, 'crop_train.csv'), index=False)
crop_test.to_csv(os.path.join(PROCESSED_DIR, 'crop_test.csv'),  index=False)

print(f'crop_df  → train: {crop_train.shape}, test: {crop_test.shape}')

# ── yield_df  (target: Yield) ────────────────────────────────────────
yield_df_clean = yield_df.dropna().reset_index(drop=True)

yield_train, yield_test = train_test_split(
    yield_df_clean, test_size=0.2, random_state=42
)

yield_train.to_csv(os.path.join(PROCESSED_DIR, 'yield_train.csv'), index=False)
yield_test.to_csv(os.path.join(PROCESSED_DIR, 'yield_test.csv'),  index=False)

print(f'yield_df → train: {yield_train.shape}, test: {yield_test.shape}')
print('\nAll files saved to:', os.path.abspath(PROCESSED_DIR))
```
